# Exercise 5: Transformer Language Model

This notebook trains a GPT-style transformer on Shakespeare text and evaluates generation quality.

```python
# Cell 1: Setup
import numpy as np
import torch
np.random.seed(42)
torch.manual_seed(42)

from transformer_solution import (
    run_training, plot_loss_curves, plot_accuracy_curves,
    display_generated_sentences, compare_generations
)
from transformer import generate_text
```

```python
# Cell 2: Q2-Q3 Training
model, data_handler, history = run_training(epochs=10, verbose=True)
```

```python
# Cell 3: Q2 Loss Plot
plot_loss_curves(history)
```

```python
# Cell 4: Q2 Accuracy Plot
plot_accuracy_curves(history)
```

```python
# Cell 5: Q3 Generated Sentences
display_generated_sentences(history)
```

```python
# Cell 6: Q4 Top-k Comparison
device = next(model.parameters()).device
block_size = 10

# Get final epoch's standard sentences
_, standard = history['generated_sentences'][-1]

# Generate with top-k=5
topk = [generate_text(model, data_handler, "the ", 30, block_size, device, top_k=5)
        for _ in range(3)]

compare_generations(standard, topk, k=5)
```
